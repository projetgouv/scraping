````python
import pandas as pd
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
from transformers import pipeline

# Charger le modèle de sentiment analysis et le tokenizer
tokenizer = AutoTokenizer.from_pretrained("tblard/tf-allocine", use_fast=True)
model = TFAutoModelForSequenceClassification.from_pretrained("tblard/tf-allocine")

# Créer le pipeline de sentiment analysis
nlp = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)

# Exemple de DataFrame avec des notes négatives
df_neg = pd.DataFrame({
    'text': [
        "Le film était vraiment mauvais, je n'ai pas du tout aimé l'histoire.",
        "Les acteurs étaient médiocres et le scénario était prévisible.",
        "La réalisation était désastreuse, je ne recommande pas ce film."
    ]
})

# Ajouter la prédiction de sentiment pour chaque texte de note négative
df_neg['sentiment'] = df_neg['text'].apply(lambda x: nlp(x)[0]['label'])

# Filtrer les textes ayant un sentiment négatif
df_neg = df_neg[df_neg['sentiment'] == 'NEGATIVE']

# Créer une liste pour stocker les mots négatifs
negative_words = []

# Tokenizer chaque texte négatif et ajouter les mots à la liste des mots négatifs
for text in df_neg['text']:
    tokens = tokenizer.tokenize(text)
    negative_words.extend(tokens)

# Compter le nombre d'occurrences de chaque mot négatif
word_counts = pd.Series(negative_words).value_counts()

# Calculer la médiane des scores
median_score = word_counts.mean()

# Récupérer les mots ayant un score supérieur à la médiane
most_negative_words = word_counts[word_counts < median_score].index.tolist()

# Afficher les mots les plus négatifs
print("Mots les plus négatifs avec un score supérieur à la médiane :")
for word in most_negative_words:
    print(word)
````

````python
import pandas as pd
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
from transformers import pipeline

# Charger le modèle de sentiment analysis et le tokenizer
tokenizer = AutoTokenizer.from_pretrained("tblard/tf-allocine", use_fast=True)
model = TFAutoModelForSequenceClassification.from_pretrained("tblard/tf-allocine")

# Créer le pipeline de sentiment analysis
nlp = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)

# Exemple de DataFrame avec des notes négatives
df_neg = pd.read_csv('/Users/camille/repo/Hetic/projet_gouv/scraping/Cleaning_eda/df_ng.csv')

# Ajouter la prédiction de sentiment pour chaque texte de note négative
df_neg['sentiment'] = df_neg['cleaned_text'].apply(lambda x: nlp(x)[0]['label'] if isinstance(x, str) else None)

# Filtrer les textes non vides ayant un sentiment négatif
df_neg_filtered = df_neg.dropna(subset=['sentiment'])
df_neg_filtered = df_neg_filtered[df_neg_filtered['sentiment'] == 'NEGATIVE']

# Créer une liste pour stocker les mots négatifs
negative_words = []

# Tokenizer chaque texte négatif et ajouter les mots à la liste des mots négatifs
for text in df_neg_filtered['cleaned_text']:
    tokens = tokenizer.tokenize(text)
    negative_words.extend(tokens)

# Compter le nombre d'occurrences de chaque mot négatif
word_counts = pd.Series(negative_words).value_counts()

# Calculer la médiane des scores
median_score = word_counts.median()

# Récupérer les mots ayant un score supérieur à la médiane
most_negative_words = word_counts[word_counts > median_score].index.tolist()

# Afficher les mots les plus négatifs
print("Mots les plus négatifs avec un score supérieur à la médiane :")
for word in most_negative_words:
    print(word)
````
